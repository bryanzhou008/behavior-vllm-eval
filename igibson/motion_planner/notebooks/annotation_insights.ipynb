```python
import os
import json
import matplotlib.pyplot as plt
import numpy as np

# Directory containing JSON files
directory = annotation_dir

# List to store lengths of lists from each JSON file
list_lengths = []

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
            # Assuming each file contains a single list
            list_lengths.append(len(data))

# Generating a histogram of the list lengths
plt.figure(figsize=(10, 6))
# Choosing an appropriate number of bins to smooth the histogram
bin_size = 2  # Adjust this value to change the smoothing level
bins = range(min(list_lengths), max(list_lengths) + bin_size, bin_size)
plt.hist(list_lengths, bins=bins, color='skyblue', edgecolor='grey')
plt.title('Length Distribution of Human Annotated Behavior-100 Tasks', fontweight='bold', fontsize=14)
plt.xlabel('Action Sequence Length', fontweight='bold')
plt.ylabel('# B-100 Tasks', fontweight='bold')
plt.grid(True)
plt.show()
```

```python
import os
import json
import matplotlib.pyplot as plt
from collections import Counter

# Directory containing the JSON files
directory_path = annotation_dir

# List to store data from all JSON files
all_data = []

# Dictionary to store the count of list lengths
length_distribution = Counter()

# Iterate through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        
        # Open and load the JSON file
        with open(file_path, 'r') as file:
            data = json.load(file)
            all_data.append(data)
            length_distribution[len(data)] += 1

# Plotting the length distribution of lists in JSON files
lengths = list(length_distribution.keys())
counts = list(length_distribution.values())



plt.figure(figsize=(10, 6))
plt.bar(lengths, counts, color='skyblue')
plt.xlabel('Length of List in JSON')
plt.ylabel('Number of JSON Files')
plt.title('Distribution of List Lengths in JSON Files')
plt.grid(True)
plt.show()
```

```python
import os
import json
import matplotlib.pyplot as plt
from collections import Counter

# Directory containing the JSON files
directory_path = annotation_dir

# Dictionary to store the count of each action
action_distribution = Counter()

# Iterate through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        
        # Open and load the JSON file
        with open(file_path, 'r') as file:
            data = json.load(file)
            for entry in data:
                action_distribution[entry['action']] += 1

# Prepare data for plotting
actions = sorted(action_distribution, key=action_distribution.get, reverse=True)
frequencies = [action_distribution[action] for action in actions]

plt.figure(figsize=(12, 8))
plt.bar(actions, frequencies, color='cadetblue')
plt.xlabel('Actions')
plt.ylabel('Frequency')
plt.title('Frequency of Each Action in JSON Data')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
import json
import matplotlib.pyplot as plt
from collections import Counter

# Directory containing the JSON files
directory_path = annotation_dir

# Dictionary to store the count of each action, with prefixes removed
action_distribution = Counter()

# Iterate through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        
        # Open and load the JSON file
        with open(file_path, 'r') as file:
            data = json.load(file)
            for entry in data:
                # Remove "LEFT_" or "RIGHT_" prefix from the action
                action = entry['action']
                if action.startswith('LEFT_'):
                    action = action[5:]  # Remove the first 5 characters
                elif action.startswith('RIGHT_'):
                    action = action[6:]  # Remove the first 6 characters
                action_distribution[action] += 1

# Prepare data for plotting
actions = sorted(action_distribution, key=action_distribution.get, reverse=True)
frequencies = [action_distribution[action] for action in actions]

plt.figure(figsize=(12, 8))
plt.bar(actions, frequencies, color='cadetblue')
plt.xlabel('Actions')
plt.ylabel('Frequency')
plt.title('Frequency of Actions in Annotated B-100 Action Sequences')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
import json
import matplotlib.pyplot as plt
from collections import Counter

# Directory containing the JSON files
directory_path = annotation_dir

# Dictionary to store the count of each action, with prefixes removed
action_distribution = Counter()

# Iterate through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        
        # Open and load the JSON file
        with open(file_path, 'r') as file:
            data = json.load(file)
            for entry in data:
                # Remove "LEFT_" or "RIGHT_" prefix from the action
                action = entry['action']
                if action.startswith('LEFT_'):
                    action = action[5:]  # Remove the first 5 characters
                elif action.startswith('RIGHT_'):
                    action = action[6:]  # Remove the first 6 characters
                action_distribution[action] += 1

# Aggregate infrequent actions into 'other actions'
for action, count in list(action_distribution.items()):
    if count < 3:
        action_distribution['other actions'] += count
        del action_distribution[action]

# Prepare data for plotting
actions = sorted(action_distribution, key=action_distribution.get, reverse=True)
frequencies = [action_distribution[action] for action in actions]

plt.figure(figsize=(12, 8))
plt.bar(actions, frequencies, color='cadetblue')
plt.xlabel('Actions', fontweight='bold', fontsize=15)
plt.ylabel('Frequency', fontweight='bold', fontsize=15)
plt.title('Frequency of Actions in Annotated Behavior-100 Action Sequences', fontweight='bold', fontsize=18)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```
